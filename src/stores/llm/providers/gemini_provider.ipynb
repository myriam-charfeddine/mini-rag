```python
def construct_prompt(prompt: str, role: str):
        return {
            "role": role,
            "parts": prompt
        }
```

```python
chat_history = [
    construct_prompt(
                prompt="if i ask about your name, tell me it's JIMIN",
                role="user"
                )
]
```

```python
chat = genai.GenerativeModel(model_name = "gemini-1.5-flash",
                                      generation_config = genai.GenerationConfig(
                                        max_output_tokens = 1000,
                                        temperature = 0.1,
                                    )
                                    ).start_chat(
                                                    history = chat_history
                                    )
```

```python
res=chat.send_message("what is you name").text
```
